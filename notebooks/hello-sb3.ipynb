```python
from stable_baselines3 import PPO
import stable_baselines3
print(f"{stable_baselines3.__version__=}")
```

```python
import gymnasium as gym
from gymnasium.wrappers import RecordEpisodeStatistics, RecordVideo
print(f"{gym.__version__=}")
```

```python
env = gym.make("CartPole-v1", render_mode="rgb_array")
```

```python
# next we'll wrap the environment in a RecordVideo, this will save the output rgb data
# into a video file for us in the ./video directory.
# it will save the file in this format: video/rl-video-episode-x.mp4
# where x is the number of simulations run, so if we run 5 simulations it will
# save 5 videos
env = RecordVideo(env, './video')
```

```python
model = PPO("MlpPolicy", env, verbose=1)

# Train the agent for 50 steps
model.learn(total_timesteps=50)

# Train the agent for 10000 steps
# model.learn(total_timesteps=10_000)
```

```python
vec_env = model.get_env()
obs = vec_env.reset()
for i in range(1000):
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, done, info = vec_env.step(action)
    vec_env.render()
    # VecEnv resets automatically
    # if done:
    #   obs = env.reset()
```
